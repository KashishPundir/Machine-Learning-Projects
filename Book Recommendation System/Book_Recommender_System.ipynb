**Types of Recommender System:**  
**1. Popularity Based:** Making system showcasing the things which are popular and trending. Like in youtube Trending section shows trending videos.

**2. Content Based:** System which showcase the things similar to the particular content. Like, If I like sci-fi movies of particular actress, Then they show the moview of that actress.

**3. Collaborative Filtering Based:** Recommends items based on the preferences of similar users.Like, Amazon’s “Customers who bought this also bought...”

**4. Hybrid Recommender System:** Combines two or more recommendation techniques (e.g., content-based + collaborative + popularity based).Like, Netflix uses hybrid models for better accuracy and personalization.

# POPULARITY BASED RECOMMENDER SYSTEM

```python
ratings_with_name=ratings.merge(books,on='ISBN')
ratings_with_name.head()
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df.head()
```

```python
Books_in_trend = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
Books_in_trend
```

```python
Books_in_trend  = Books_in_trend.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
Books_in_trend
```

# COLLABORATIVE FILTERING BASED RECOMMENDER SYSTEM

```python
ratings_with_name
```

```python
# Count of rating contribution to books by users
ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
# Users having min 200 rating contribution to books
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
# Similarity score between each book with one another..
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores=cosine_similarity(pt)
similarity_scores.shape
```

```python
# finding index of the searched book
np.where(pt.index=='1984')
```

```python
np.where(pt.index=='1984')[0][0]
```

```python
# Simiarity of 0th book with other books (index or row,similarity with 0th row)
sorted(list(enumerate(similarity_scores[0])),key=lambda x:x[1],reverse=True)
```

```python
sorted(list(enumerate(similarity_scores[0])),key=lambda x:x[1],reverse=True)[1:6]
```

```python
def recommend(book_name):
  # index fetch
  index=np.where(pt.index==book_name)[0][0]
  similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

  for i in similar_items:
    print(pt.index[i[0]])
```
